1\. Using `np.linspace()` generate a numpy array from 0 to 10 with 1000 intermediate values, then calculate sine and cosine on that array.

2\. Do the following using the data from above:
- Plot just the sine and just the cosine
- Plot both the sine and the cosine on the same graph
- Plot the sine on the top subplot and the cosine on the bottom subplot
[//]: # (Hello) 


```python
array = np.linspace(0, 10, 1000)
array_sine = np.sin(array)
array_cosine = np.cos(array)
plt.plot(array_sine)
plt.plot(array_cosine)


figure = plt.figure()
top_subplot = figure.add_subplot(2,1,1)
top_subplot.plot(array_sine)
bottom_subplot = figure.add_subplot(2,1,2)
bottom_subplot.plot(array_cosine, color = "green")

```

4\. Using the same data and figures from question 1, plot both sin(x) and cos(x) on the same graph, then do the following:

- Set a legend, title, x-label, and y-label 

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.plot(array_sine)
plt.plot(array_cosine)

ax.set_title('Graph of Sine and Cosine Functions')
ax.set_xlabel('Distance')
ax.set_ylabel('Height')
ax.legend(labels = ['Sine','Cosine'])
```
